```python
#使用するデータを使用する際に貼る必要がありそうだったurl http://www3.dsi.uminho.pt/pcortez/student.pdf
import pandas as pd
students = pd.read_csv("student-por.csv",sep=";",encoding="utf-8")
students
```

```python
#27,Dalc 28 Walc 平日、休日のアルコール消費量　29 health 健康　全て１～５の数字

al={}

#D W の値を_で分け、それぞれのグループで平均のhealth値を求める
for a,st in students.iterrows():
    D,W,h = (int(st["Dalc"]),int(st["Walc"]),float(st["health"]))
    key = str(D)+ "_" +str(W)
    if not (key in al):
        al[key] = []
    al[key] += [h]
al= sorted(al.items())
al = dict((x,y) for x,y in al)
#print(al)
avs = {}

for key in al:
    avs[key]= sum(al[key])/len(al[key])
    v=avs[key]
    print("{0}  {1}".format(key,v))

#print(avs)
sortedaves=sorted(avs.items(),key=lambda x:x[1])
#print(sortedaves)

print("")

for i,j in sortedaves:
    print("{0} {1}".format(i,j))
#相関はなさそう
```
